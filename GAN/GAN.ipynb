```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import torchvision

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

class Generator(nn.Module):
    def __init__(self):
        super(Generator, self).__init__()
        self.main = nn.Sequential(
            nn.Linear(100, 256),
            nn.ReLU(),
            nn.Linear(256, 512),
            nn.ReLU(),
            nn.Linear(512, 28 * 28),
            nn.Tanh()
        )

    def forward(self, x):
        return self.main(x)

class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()
        self.main = nn.Sequential(
            nn.Linear(28 * 28, 128),
            nn.Softplus(),
            nn.Linear(128, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        return self.main(x)

transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

train_dataset = datasets.MNIST(root='./data', train=True, transform=transform)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_set = datasets.MNIST(root='./data', train=False, transform=transform)
test_loader = DataLoader(test_set, batch_size=32, shuffle=False)

G = Generator().to(device)
D = Discriminator().to(device)


criterion = nn.BCELoss()
optimizer_G = optim.Adam(G.parameters(), lr=0.0001)
optimizer_D = optim.Adam(D.parameters(), lr=0.0001)

```

```python
# Add lists to store accuracies for each epoch
num_epochs = 100
D_real_acc_train = []
D_fake_acc_train = []
D_real_acc_test = []
D_fake_acc_test = []
z = torch.randn(16, 100).to(device)
fake_images = G(z)
images = torch.zeros(16, 1,28,28, num_epochs +1)
fake_images = fake_images.view(fake_images.size(0), 1, 28, 28)
images[:,:,:,:,0] = fake_images
# Add lists to store Losses for each epoch
G_losses = []
D_losses = []
for epoch in range(num_epochs):
    real_correct = 0
    fake_correct = 0
    total_samples = 0
    G_loss = 0
    D_loss = 0

    for batch_idx, (real_images, _) in enumerate(train_loader):
        D.train()
        G.train()
        batch_size = real_images.size(0)
        total_samples += batch_size

        # Discriminator with real images
        real_images = real_images.view(batch_size, -1).to(device)
        real_labels = torch.ones(batch_size, 1).to(device)
        fake_labels = torch.zeros(batch_size, 1).to(device)

        optimizer_D.zero_grad()
        outputs = D(real_images)
        d_loss_real = criterion(outputs, real_labels)

        # Accuracy for real images
        real_predictions = outputs > 0.5
        real_correct += real_predictions.eq(real_labels).sum().item()

        # Discriminator with fake images
        z = torch.randn(batch_size, 100).to(device)
        fake_images = G(z)
        outputs = D(fake_images.detach())
        d_loss_fake = criterion(outputs, fake_labels)

        # Accuracy for fake images
        fake_predictions = outputs < 0.5
        fake_correct += fake_predictions.eq(fake_labels).sum().item()

        d_loss = d_loss_real + d_loss_fake
        d_loss.backward()
        optimizer_D.step()

        # Generator update
        optimizer_G.zero_grad()
        z = torch.randn(batch_size, 100).to(device)
        fake_images = G(z)
        outputs = D(fake_images)
        g_loss = criterion(outputs, real_labels)
        g_loss.backward()
        optimizer_G.step()
        G_loss += g_loss.item()
        D_loss += d_loss.item()

    # Calculate accuracy for this epoch
    real_accuracy = 100 * real_correct / total_samples
    fake_accuracy = 100 * fake_correct / total_samples
    D_real_acc_train.append(real_accuracy)
    D_fake_acc_train.append(fake_accuracy)
    G_losses.append(G_loss)
    D_losses.append(D_loss)

    print(f'Epoch [{epoch+1}/{num_epochs}], d_loss: {d_loss.item():.4f}, g_loss: {g_loss.item():.4f}, '
          f'Real Acc: {real_accuracy:.2f}%, Fake Acc: {fake_accuracy:.2f}%')
    
    real_correct = 0
    fake_correct = 0
    total_samples = 0
    for batch_idx, (real_images, _) in enumerate(test_loader):
        batch_size = real_images.size(0)
        total_samples += batch_size
        # Discriminator with real images
        real_images = real_images.view(batch_size, -1).to(device)
        real_labels = torch.ones(batch_size, 1).to(device)
        fake_labels = torch.zeros(batch_size, 1).to(device)
        
        D.eval()
        G.eval()
        with torch.no_grad():
            outputs = D(real_images)
            real_predictions = outputs > 0.5
            real_correct += real_predictions.eq(real_labels).sum().item()
            z = torch.randn(batch_size, 100).to(device)
            fake_images = G(z)
            outputs = D(fake_images.detach())
            # Accuracy for fake images
            fake_predictions = outputs < 0.5
            fake_correct += fake_predictions.eq(fake_labels).sum().item()
    # Calculate accuracy for this epoch
    real_accuracy = 100 * real_correct / total_samples
    fake_accuracy = 100 * fake_correct / total_samples
    D_real_acc_test.append(real_accuracy)
    D_fake_acc_test.append(fake_accuracy)
        
        
    fake_images = fake_images.view(fake_images.size(0), 1, 28, 28)
    images[:,:,:,:,epoch + 1] = fake_images
    if (epoch+1) % 10 == 0:
        fake_images = fake_images.view(fake_images.size(0), 1, 28, 28)
        images[:,:,:,:,epoch] = fake_images
        grid = torchvision.utils.make_grid(fake_images[:25], nrow=5, normalize=True)
        plt.imshow(grid.permute(1, 2, 0).cpu())
        plt.axis('off')
        plt.show()

# Plot Discriminator accuracy over epochs
plt.figure(figsize=(10,5))
plt.plot(D_real_acc_train, label='Real Image Accuracy (Train)')
plt.plot(D_fake_acc_train, label='Fake Image Accuracy (Train)')
plt.plot(D_real_acc_test, label='Real Image Accuracy (Test)')
plt.plot(D_fake_acc_test, label='Fake Image Accuracy (Test)')
plt.title('Discriminator Accuracy over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuracy (%)')
plt.legend()
plt.show()

plt.figure(figsize=(10,5))
plt.plot(G_losses, label='Generator Loss')
plt.plot(D_losses, label='Discrimanator Loss')
plt.title('Loss over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Step 2: Set up the figure and axis for animation
fig, axs = plt.subplots(4, 4, figsize=(16, 8))
plt.subplots_adjust(hspace=0.5, wspace=0.5)
images = images.detach().cpu()
# Turn off axes for all subplots
for ax in axs.flatten():
    ax.axis('off')

# Step 3: Function to update the frame for the animation
def update(frame):
    for i in range(4):
        for j in range(4):
            idx = i * 4 + j
            if idx < images.shape[0]:
                axs[i, j].imshow(images[idx, :, :, frame], cmap='gray', interpolation='nearest')
            else:
                axs[i, j].imshow(np.zeros((28, 28)), cmap='gray')  # Blank for empty subplots
    print(frame)
    return axs.flatten()

# Step 4: Create the animation
ani = FuncAnimation(fig, update, frames=101, blit=False)

# Step 5: Save the animation
ani.save('animation.gif', writer='pillow', fps=10)

# Optional: Show the animation
plt.show()
```
